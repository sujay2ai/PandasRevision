```python
df1 = pd.read_csv(io.BytesIO(uploaded['cars.csv']),skiprows = 1)  #To skip rowsfrom top -- this will skip 1 row from top -- Skipped the header in this case.
# ---- OR----
#df1 = pd.read_csv(io.BytesIO(uploaded['cars.csv']), header = 1) # OR you can say header = 1 -- Means your header is located at row-2 (as index starts from 0)
df1
```

```python
df2 = pd.read_csv(io.BytesIO(uploaded['cars.csv']),header = None)  # If you fie does not contain the header. Automatically column names are generated like 1,2,3 etc
df2
```

```python
df3 = pd.read_csv(io.BytesIO(uploaded['cars.csv']),header = None, names = ["Model" , "mpg", "cyl", "Disp" , "hp"]) 
# If your file does not contain the header. You can specify the column names as an argument.
df3
```

```python
df4 = pd.read_csv(io.BytesIO(uploaded['cars.csv']),nrows = 5) # If you want to view only few rows -- Specify number of rows -- It will read no. of rows excluding the Header.
df4
```

```python
df5 = pd.read_csv(io.BytesIO(uploaded['cars.csv']),na_values=["n.a", "not available"]) 
# If any cell contains value "n.a" or "not available", then convert them to NaN value. -- Used in Cleaning Dats.
# We cannot include -1 in the na_values list, as we do not want to convert -1 to NaN for all columns, but just few of them.
df5
```

```python
# We cannot include -1 in the na_values list, as we do not want to convert -1 to NaN for all columns, but just few of them.
# For this, we can use a dictionary instead of List.
df6 = pd.read_csv(io.BytesIO(uploaded['cars.csv']),na_values={
    'model' : ["n.a", "not available", -1],
    'mpg' : ["n.a", "not available"],
    'cyl' : ["n.a", "not available",-1],
    'disp' : ["n.a", "not available"],
    'hp' : ["n.a", "not available"]
})  # if we want to eliminate -1 from few columns. -- We can mention column wise, the values we want to convert to NaN value.
# Data Munging / Data Wrangling / Data Cleaning.
df6
```

                                                      
                                                      -----WRITING TO CSV-----

```python
df6.to_csv("new.csv")   # To write to CSV.
```

```python
df6.to_csv("new.csv",index = false)   # If we do not want to write the index in CSV.
```

```python
df6.to_csv("new.csv",columns = ['Model','hp'])   # If we want to write only selected columns to CSV.
```

```python
df6.to_csv("new.csv",headers = False)   # If user wants to skip writing the headers to CSV.
```

```python
import pandas as pd
from google.colab import files  #To read file from local
import io
uploaded = files.upload()   # To upload file from local.

def convert_mpg_cell(cell):    # Function to convert values of particular column.  --- If value is found to be "not available", then replace with 25.
  if cell == "not available":
    return 25
  return cell

df8 = pd.read_excel(io.BytesIO(uploaded['cars.xlsx']),"Sheet1", converters = {
    'mpg' : convert_mpg_cell         #Calling function.  --- Similarly can be done for other columns -- Write different functions for different columns. 
})   
```

```python
df8
```

```python
df8.to_excel("new.xlsx" , sheet_name = "cars_sheet")   # to write to excel -- mention sheet name.
```

```python
df8.to_excel("new.xlsx" , sheet_name = "cars_sheet", startrow = 2, startcol = 3)   # to start writing from particular row / column.
```

```python
df8.to_excel("new.xlsx" , sheet_name = "cars_sheet", index = False)   # If want to skip writing index column.
```

```python
# Writing 2 dataframes in same excel but in different sheet.

df_stocks = pd.DataFrame({                   # 1st Dataframe.
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})
df_weather =  pd.DataFrame({                 # 2nd Dataframe
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
